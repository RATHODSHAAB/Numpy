# Broadcasting

The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.
The smaller array is broadcast across the larger array so that they have compatible shapes.

Broadcasting is done by the numpy itself ! we just know will broadcasting happen or not ? that's it !
To perform broadcasting you have to follow a set of 3 rules, <br/>
Rule 1: <br/>
**Make the two arrays have the same number of dimensions.**
If the numbers of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the smaller dimension.

Rule 2 : <br/>
**Make each dimension of the two arrays the same size.**

If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size of the other array.<br/>
Rule 3:<br />
**If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an error is raised.**

```python
a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)
print(b)

print(a+b)
# This will give result because broadcasting will happen here!
# b array will be converted into 2D as it is in 1D by adding 1 in head (1,3) : now this one will be
# stretched to 4[corresponding to array a.] array A (4,3) =  Array B(4,3)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)

print(a)
print(b)

print(a + b)
# Same this will also perform broadcasting
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

print(a)
print(b)


# There will be no bradcasting as per 3rd rules. print(a+b)
```
